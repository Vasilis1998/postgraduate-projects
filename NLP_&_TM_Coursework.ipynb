# Data Preprocessing

```python
# Handling missing values

# Drop rows with missing product UID
attributes_data.dropna(subset=['product_uid'], inplace=True)

# Fill missing values in the name column with an empty string
attributes_data['name'].fillna('', inplace=True)

# Fill missing values in the value column with a predefined value
attributes_data['value'].fillna('Unknown', inplace=True)
```

```python
# So now there are not missing values
print(attributes_data.isnull().sum())
```

```python
# Removing special characters, puncutation marks and non-alphanumeric symbols

import re

def remove_special_characters(text):
    # Remove special characters, puncutation marks and non-alphanumeric symbols
    cleaned_text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return cleaned_text

# I apply the function to training data
train_data['product_title'] = train_data['product_title'].apply(remove_special_characters)
train_data['search_term'] = train_data['search_term'].apply(remove_special_characters)

# I apply the function to product description data
product_desc_data['product_description'] = product_desc_data['product_description'].apply(remove_special_characters)

# I apply the function to attributes data
attributes_data['name'] = attributes_data['name'].apply(remove_special_characters)
attributes_data['value'] = attributes_data['value'].apply(remove_special_characters)
```

```python
# Lowercasing

# I apply lowercase to training data
train_data['product_title'] = train_data['product_title'].str.lower()
train_data['search_term'] = train_data['search_term'].str.lower()

# I apply lowercase to product description data
product_desc_data['product_description'] = product_desc_data['product_description'].str.lower()

# I apply lowercase to attributes data
attributes_data['name'] = attributes_data['name'].str.lower()
attributes_data['value'] = attributes_data['value'].str.lower()
```

```python
# Remove stopwords

import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords

stopwords = set(stopwords.words('english'))

def remove_stopwords(text):
    # Tokenize the text into words
    words = nltk.word_tokenize(text)
    
    # Remove stopwords from the list of words
    words = [word for word in words if word.lower() not in stopwords]
    
    # Join the remaining words back into a single string
    cleaned_text = ' '.join(words)
    return cleaned_text

# I remove the stopwords from training data
train_data['product_title'] = train_data['product_title'].apply(remove_stopwords)
train_data['search_term'] = train_data['search_term'].apply(remove_stopwords)

# I remove the stopwords from product description data
product_desc_data['product_description'] = product_desc_data['product_description'].apply(remove_stopwords)

# I remove the stopwords from attributes data
attributes_data['name'] = attributes_data['name'].apply(remove_stopwords)
attributes_data['value'] = attributes_data['value'].apply(remove_stopwords)
```

```python
# Lemmatization

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
lemmatizer = WordNetLemmatizer()

# Define a function to apply lemmatization to a text
def lemmatize_text(text):
    tokens = nltk.word_tokenize(text)  # Tokenize the text into words
    lemmas = [lemmatizer.lemmatize(token) for token in tokens]  # Lemmatize each word
    lemmatized_text = ' '.join(lemmas)  # Join the lemmatized words back into a single string
    return lemmatized_text

# Apply lemmatization to the 'product_title' column in the train_data 
train_data['product_title'] = train_data['product_title'].apply(lemmatize_text)

# Apply lemmatization to the 'search_term' column in the train data
train_data['search_term'] = train_data['search_term'].apply(lemmatize_text)

# Apply lemmatization to the 'product_description' column in the product description data
product_desc_data['product_description'] = product_desc_data['product_description'].apply(lemmatize_text)

# Apply lemmatization to the 'name' column in the attributes data
attributes_data['name'] = attributes_data['name'].apply(lemmatize_text)

# Apply lemmatization to the 'value' column in the attributes data
attributes_data['value'] = attributes_data['value'].apply(lemmatize_text)
```

```python
# Handling misspelled words

!pip install autocorrect
from autocorrect import Speller

# Initialize the spell checker
spell = Speller()

# Define a function to correct misspelled words in a text
def correct_spelling(text):
    # Tokenize the text into words
    words = text.split()
    
    # Iterate through each word
    corrected_words = []
    for word in words:
        # Check if the word is misspelled
        corrected_word = spell(word)
        corrected_words.append(corrected_word)
    
    # Join the corrected words back into a single string
    corrected_text = ' '.join(corrected_words)
    return corrected_text

# Apply spelling correction to the 'search_term' column in the train_data
train_data['search_term'] = train_data['search_term'].apply(correct_spelling)
```

# Feature Extraction

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

# We apply word embeddings using Word2Vec to convert the textual data into numerical representations

def apply_word2vec(text_column, vector_size=100, window=5, min_count=1, workers=4):
    # Tokenize the text in the column
    tokenized_text = text_column.apply(word_tokenize)
    
    # Train Word2Vec model on the tokenized text
    model = Word2Vec(tokenized_text, vector_size=vector_size, window=window, min_count=min_count, workers=workers)
    
    return model

# Apply Word2Vec to each column
product_title_model = apply_word2vec(train_data['product_title'])
search_term_model = apply_word2vec(train_data['search_term'])
product_description_model = apply_word2vec(product_desc_data['product_description'])
name_model = apply_word2vec(attributes_data['name'])
value_model = apply_word2vec(attributes_data['value'])
```

# Models 

```python
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import xgboost as xgb
from sklearn.metrics import mean_squared_error

# Split the data into training and testing sets
train_x, test_x, train_y, test_y = train_test_split(train_data, train_data['relevance'], test_size=0.3, random_state=42)
```

```python
# Define a list of models with their respective parameter grids
models = [
    (LinearRegression(), {}),
    (RandomForestRegressor(), {'n_estimators': [10, 50, 100]}),
    (xgb.XGBRegressor(), {'learning_rate': [0.1, 0.01, 0.001]}),
    (GradientBoostingRegressor(), {'n_estimators': [10, 50, 100]}),
    (Lasso(), {'alpha': [0.1, 0.01, 0.001]}),
    (Ridge(), {'alpha': [0.1, 0.01, 0.001]})
]
```

```python
# Function to retrieve the embedding of a search term from the Word2Vec model
def get_embedding(term):
    try:
        return search_term_model.wv[term]
    except KeyError:
        # Handle missing search terms by returning a default embedding or None
        return np.zeros(search_term_model.vector_size)  # You can choose a default value or handle missing terms differently

# Extract the numerical representations of search terms
train_x_emb = train_x['search_term'].apply(lambda x: get_embedding(x))
test_x_emb = test_x['search_term'].apply(lambda x: get_embedding(x))

# Convert the embeddings to numpy arrays
train_x_emb = np.vstack(train_x_emb)
test_x_emb = np.vstack(test_x_emb)
```

```python
# Initialize variables to store the best model and its RMSE
best_model = None
best_rmse = float('inf')

# Iterate through each model, perform grid search, and calculate RMSE
for model, param_grid in models:
    grid_search = GridSearchCV(model, param_grid, scoring='neg_mean_squared_error')
    grid_search.fit(train_x_emb, train_y)
    
    # Get the best model and its parameters
    best_estimator = grid_search.best_estimator_
    best_params = grid_search.best_params_
    
    # Fit the best model
    best_estimator.fit(train_x_emb, train_y)
    pred_y = best_estimator.predict(test_x_emb)
    rmse = np.sqrt(mean_squared_error(test_y, pred_y))
    
    print("Applied Model: {}, Best Parameters: {}, Root Mean Squared Error: {}".format(type(model).__name__, best_params, rmse))
    
    # Update the best model if the current RMSE is lower
    if rmse < best_rmse:
        best_model = best_estimator
        best_rmse = rmse
```

```python
# Print the best model and its RMSE
print("Best Model: {}, Lowest Root Mean Squared Error: {}".format(type(best_model).__name__, best_rmse))
```
